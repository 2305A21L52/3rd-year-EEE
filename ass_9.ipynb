```python
import numpy as np

marks=np.array([[[164,87,96,78,83,100,100],

                 [165,89,87,76,90,79,87],

                 [164,87,100,99,100,98,100],

                 [151,15,36,76,63,24,57],

                 [165,96,91,82,76,72,87]],

                 [[120,78,89,92,69,98,86],

                 [151,19,56,67,76,91,65],

                 [152,72,84,83,77,98,89],

                 [120,67,89,73,95,68,86],

                 [152,99,89,92,97,87,79]]])

stud=marks.shape[1]
sem=marks.shape[0]

Hall_Ticket_No=int(input("enter hall ticket no:"))

semester=int(input("enter semester no:"))

count=0
failed_sub=[]
for i in range(sem):

   if(semester==i+1):

      break

index_sem=i

for j in range(stud):

  if (Hall_Ticket_No==marks [index_sem,j,0]):

    break

index_stud=j


marks_percentage=np.sum(marks [index_sem, index_stud, 1: marks.shape[2]])*100/600

for i in range(1,marks.shape[2]):

  if (marks [index_sem, index_stud, i]<40):

     count=count+1

     failed_sub.append(i)

if(count>0):

    print("Fail")

    print("Failed Subjects:",)

    for i in range(len(failed_sub)):

        if(failed_sub[i]==1):

            print("Subject1")

        if(failed_sub[i]==2):

            print("Subject2")

        if(failed_sub[i]==3):

            print("Subject3")

        if(failed_sub[i]==4):

            print("Subject4")

        if(failed_sub[i]==5):

            print("Subject5")

        if(failed_sub[i]==6):

            print("Subject6")

else:

    print("=============================")

    print("       SR University          ")

    print("           CGPA               ")

    print("=============================")

    print("Roll No: ", Hall_Ticket_No)

    print("Semester:", semester)

    print("Subject1:", marks [index_sem,index_stud, 1])

    print("Subject2:", marks [index_sem,index_stud,2])

    print("Subject3:", marks [index_sem,index_stud,3])

    print("Subject4:", marks [index_sem,index_stud,4])

    print("Subject5:", marks [index_sem,index_stud,5])

    print("Subject6:", marks [index_sem,index_stud,6])

    print("=============================")

    print("          Results            ")
    print("=============================")
    print("Pass")
    print("Percentage of marks:",np.round(marks_percentage))

    print("============================")


for i in range(1,marks.shape[2]):

  if (marks [index_sem, index_stud, i]<40):

     count=count+1

     failed_sub.append(i)


```
